Credit Card Fraud Detection using CRISP-DM

Business Understanding

Problem Statement:

Financial institutions face significant losses due to fraudulent credit card transactions. Fraudulent transactions are rare but highly costly.

Business Question:
Can we build a predictive model that accurately detects fraudulent transactions while minimizing false negatives (missed fraud cases)?

Business Objective:

Maximize Recall for fraud class

Reduce financial loss

Maintain reasonable precision to avoid customer inconvenience

2. Data Understanding

We use a publicly available Credit Card Fraud dataset.
The dataset contains anonymized features (V1–V28), transaction amount, and a target variable (Class).

Class = 0 → Normal Transaction

Class = 1 → Fraud Transaction
